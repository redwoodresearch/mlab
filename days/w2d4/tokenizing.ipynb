```python
def corpus_common_tokens(corpus):
    corpus = [re.findall(r"\w+|[^\w\s]", corpus_el) for corpus_el in corpus]
    corpus = [item for sublist in corpus for item in sublist]
    corpus = np.array(corpus)
    unique_tokens, indices, counts = np.unique(corpus, return_counts=True, return_index= True)
    return list(a[indices[np.argsort(-counts)]])[:30000]
```
